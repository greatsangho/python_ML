```python
import faiss
from sentence_transformers import SentenceTransformer
import numpy as np
```

```python
# 임베딩모델 로드
model =  SentenceTransformer('all-MiniLM-L6-v2')
```

```python
# 텍스트 데이터 벡터화
import numpy as np
text = np.array(['강아지는 충성 스럽다','강아지는 귀엽다','강아지는 반려동물이다','개는 사람을 좋아한다.','개는 주인을 지킨다'])
text_embedding = model.encode(text)
text_embedding
```

```python
# 수치데이터를 가지고 벡터 데이터베이스 구축  FAISS
dimention =  text_embedding.shape[1]  # 벡터 차원
faiss_index = faiss.IndexFlatL2(dimention)  # 벡터 데이터 베이스 초기화  L2:유클리드 거리 기반의 유사도 방식
faiss_index.add(text_embedding) # 저장
```

```python
# 쿼리 텍스트 임베딩, 유사도 계산
query = '사랑스러운'
query_embedding = model.encode([query])
D,I =  faiss_index.search(query_embedding,k=2)  # 상위 2  D:유사도 점수, I:배열
text[I]
```

```python
# https://drive.google.com/file/d/1aU4FWR61LCoNXfYJzfrh4D9xi55gmu9b/view?usp=sharing
import gdown
import pandas as pd

file_id = '1aU4FWR61LCoNXfYJzfrh4D9xi55gmu9b'
download_url = f'https://drive.google.com/uc?id={file_id}'
gdown.download(download_url, 'chatbotdata.csv', quiet=False)
df = pd.read_csv('chatbotdata.csv')
datas = df[['Q','A']].to_numpy()
datas[:3]
```

```python
import torch
import torch_directml
from transformers import GPT2LMHeadModel
from transformers import PreTrainedTokenizerFast

tokenizer = PreTrainedTokenizerFast.from_pretrained("skt/kogpt2-base-v2",
bos_token='</s>', eos_token='</s>', unk_token='<unk>',
pad_token='<pad>', mask_token='<mask>')

model = GPT2LMHeadModel.from_pretrained('skt/kogpt2-base-v2')
tokenized_inputs = []
for data in datas[:200]:  # 시간관계상 200개만 사용
  question = data[0]
  answer = data[1]
  input_text = f"질문:{question} 답:{answer}"
  input_ids = tokenizer(input_text, truncation=True, padding='max_length', max_length = 512,  return_tensors='pt')
  tokenized_inputs.append(input_ids)
# 모델 학습
# 데이터셋 과 학습루프를 이용해서 학습
from torch.utils.data import Dataset, DataLoader
class QADataset(Dataset):
  def __init__(self, tokenized_inputs):
    self.tokenized_inputs = tokenized_inputs
  def __len__(self):
    return len(self.tokenized_inputs)
  def __getitem__(self, idx):
    return self.tokenized_inputs[idx]
train_dataset = QADataset(tokenized_inputs)
train_dataloader = DataLoader(train_dataset, batch_size=4, shuffle=True)
# 학습루프..... 옵티마이져, 스케줄러
from transformers import AdamW
optimizer = AdamW(model.parameters(), lr=1e-5)
from transformers import get_linear_schedule_with_warmup
epochs = 5
total_steps = len(train_dataloader) * epochs
scheduler = get_linear_schedule_with_warmup(optimizer, num_warmup_steps=0, num_training_steps=total_steps)
# 디바이스 선택
# device = torch.device('cuda') if torch.cuda.is_available() else torch.device('cpu')
device = torch_directml.device()
model.to(device)
from tqdm import  tqdm
for epoch in range(epochs):
  model.train()
  iterator = tqdm(train_dataloader)
  for batch in iterator:
    input_ids = batch['input_ids'].to(device)
    attention_mask = batch['attention_mask'].to(device)
    # 예측
    outputs = model(input_ids, attention_mask=attention_mask, labels=input_ids)
    loss = outputs.loss
    loss.backward()
    optimizer.step()
    scheduler.step()
    optimizer.zero_grad()
    iterator.set_description(f"Epoch {epoch}")
    iterator.set_postfix({'loss': f'{loss.item():.4f}'})

torch.save(model.state_dict(), './model/kogpt2_qa.pth')


# input_ids = tokenizer.encode(text, return_tensors='pt')
# gen_ids = model.generate(input_ids,
#   max_length=128,
#   repetition_penalty=2.0,
#   pad_token_id=tokenizer.pad_token_id,
#   eos_token_id=tokenizer.eos_token_id,
#   bos_token_id=tokenizer.bos_token_id,
#   use_cache=True)
# generated = tokenizer.decode(gen_ids[0])
# print(generated)
```
