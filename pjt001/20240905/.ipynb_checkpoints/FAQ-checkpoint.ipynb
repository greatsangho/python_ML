```python
import urllib
from bs4 import BeautifulSoup
import re
```

```python
url = 'https://www.car365.go.kr/web/contents/etc_faq_compl.do?&schM=list&page=1&viewCount=100&id=&schBdcode=&schGroupCode='
html = urllib.request.urlopen(url)
soup = BeautifulSoup(html,'html.parser')



# def FAQ_1():
#     categori = []
#     question = []
#     answer = []
#     c = soup.select('#wrap > div > div.contentsWrap > div.sub_inner > div:nth-child(7) > div > div.notice_board.notice_mob > table > tbody > tr')
#     for i in range(0,len(c),2):
#         k, q = c[i].text.strip().split('\n')[0],c[i].text.strip().split('\n')[1]
#         a = c[i+1].text.strip()
#         # re_a = re.findall(r'[가-힣0-9\s.]+',a)
#         # an = ' '.join(re_a).replace('\xa0','').replace('\n','').replace('\t','')
#         categori.append(k)
#         question.append(q)
#         answer.append(a)



category = []
question = []
answer = []
c = soup.select('div.notice_board.notice_mob > table > tbody > tr')

# 데이터가 제대로 선택되었는지 확인
for i in range(0, len(c), 2):
    k, q = c[i].text.strip().split('\n')[0], c[i].text.strip().split('\n')[1]
    a = c[i+1].text.strip()

    # 정규 표현식 개선: 숫자, 한글, 공백, 특수 문자 등을 더 많이 포함
    re_a = re.findall(r'[가-힣0-9\s\.\,\?\!]+', a)
    an = ' '.join(re_a).replace('\xa0', '').replace('\n', '').replace('\t', '')
    
    category.append(k)
    question.append(q)
    answer.append(an)
        
```

```python
import pandas as pd

data = {
    'category': category,
    'question' : question,
    'answer': answer}

df = pd.DataFrame(data)
```

```python
from sqlalchemy import create_engine
username = 'root'
password = '1234'
host = 'localhost'
database = 'mydatabase'

engine = create_engine(f'mysql+pymysql://{username}:{password}@{host}/{database}')

df.to_sql('qna', con=engine, if_exists='append', index=False)

engine.dispose()
```
