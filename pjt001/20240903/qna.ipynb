```python
# json형식의 반환값을 json 객체로 변환
import pandas as pd
import os
import sys
import urllib.request as ur
import json
from bs4 import BeautifulSoup  # pip install bs4
import xml.etree.ElementTree as ET
```

```python
result = []
for i in range(1,6):
    url = f'https://www.jobkorea.co.kr/help/faq/user?tab={i}'
    html = ur.urlopen(url)
    soup = BeautifulSoup(html,'html.parser')
    li_lists = tr_lists = soup.select('#content > div > div.faqWrap > div.tabSearchList > ul > li')

    for data in li_lists:
        temp = data.select('span.tabSearchListTitleWrap')[0].text.split(']')
        temp[0]=temp[0][1:]
        print(temp)
        result.append(temp)
```
