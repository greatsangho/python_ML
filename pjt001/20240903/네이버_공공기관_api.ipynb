```python
url = 'https://openapi.gg.go.kr/CarRegistration'
response = requests.get(url)

xml_string = response.content.decode('utf-8')

# XML 파싱
root = ET.fromstring(xml_string)

# XML 데이터를 파싱하여 딕셔너리로 변환하는 함수
def xml_to_dict(element):
    if len(element) == 0:  # 자식 요소가 없는 경우
        return element.text
    return {child.tag: xml_to_dict(child) for child in element}

# 루트 요소를 딕셔너리로 변환
response_dict = xml_to_dict(root)

# 딕셔너리를 JSON 객체로 변환
json_object = json.dumps(response_dict, ensure_ascii=False, indent=4)

# 결과 출력
print(json_object)
```

```python
# data = df3.values[0]
def makeRow(data):
    result = ''
    for idx,value in enumerate(data):
        if idx > 2:
             result += value+','
        else:        
             result += f"'{value}',"
    result = result[:-1]
    result = f"({result})"
    return result
rows = ''
for data in df3.values:
    rows += makeRow(data)+','
rows = rows[:-1]    
```
