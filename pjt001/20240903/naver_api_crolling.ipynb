```python
# 네이버 검색 API 예제 - 블로그 검색
import os
import sys
import urllib.request
client_id = "client_id"
client_secret = "secret"
encText = urllib.parse.quote("크롤링") # 인코딩하기
# url = "https://openapi.naver.com/v1/search/blog?query=" + encText # JSON 결과
# url = "https://openapi.naver.com/v1/search/blog.xml?query=" + encText # XML 결과
display = 100
params = f'?query={encText}&display={display}'
url = "https://openapi.naver.com/v1/search/news.json" + params # 뉴스 JSON
# ?앞까지는 서버에 접속, 함수 호출하기, ?뒷쪽은 파라미터
request = urllib.request.Request(url) # request 사용
request.add_header("X-Naver-Client-Id",client_id) # 헤더정보 세팅
request.add_header("X-Naver-Client-Secret",client_secret)
response = urllib.request.urlopen(request)
rescode = response.getcode()
if(rescode==200): # 성공시 반환하는 코드
    response_body = response.read()
    print(response_body.decode('utf-8'))
else:
    print("Error Code:" + rescode)
```

```python
import xml.etree.ElementTree as ET

# 주어진 XML 데이터 (바이너리 문자열을 디코딩하여 문자열로 변환)
xml_data_bytes = response

# 바이너리 문자열을 UTF-8로 디코딩
xml_data = xml_data_bytes.content.decode('utf-8')

# XML 데이터 파싱
root = ET.fromstring(xml_data)

# XML 데이터를 딕셔너리로 변환하는 함수
def xml_to_dict(element):
    if len(element) == 0:
        # Leaf 노드인 경우
        return element.text.strip() if element.text else None
    
    # 자식 요소를 딕셔너리로 변환
    result = {}
    for child in element:
        child_dict = xml_to_dict(child)
        if child.tag in result:
            # 동일한 태그가 여러 번 있는 경우
            if not isinstance(result[child.tag], list):
                result[child.tag] = [result[child.tag]]
            result[child.tag].append(child_dict)
        else:
            result[child.tag] = child_dict
    
    return result

# 루트 요소를 딕셔너리로 변환
xml_dict = xml_to_dict(root)

# # 결과 출력
# import pprint
# pprint.pprint(xml_dict)
xd = xml_dict['body']['items']['item']
```

```python
import xml.etree.ElementTree as ET

def xml_to_dict(element):
    """Recursively convert XML elements to a dictionary."""
    def parse_element(e):
        # Convert child elements
        result = {}
        for child in e:
            if len(child) > 0:  # If the child element has children
                result[child.tag] = parse_element(child)  # Recursively parse the child
            else:
                result[child.tag] = child.text  # Get the text content of the child
        return result

    return parse_element(element)
```
