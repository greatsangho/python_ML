```python
def reportNa(df):
    t = df.isna().sum()
    t[t>0]
    df_na = pd.DataFrame(t[t>0],columns=['na'])
    df_na['type'] = df.dtypes[df_na.index]
    return df_na
```

```python
df = sns.load_dataset('titanic')
reportNa(df)
```
