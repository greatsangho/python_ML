```python
# 행(row) 열(column)
# DataFrame 안에 dictionary 객체 입력
# 데이터 프레임 만들기
keys = ['c'+str(i) for i in range(5)]
li = []
values = []
for i in range(1,16):
    li.append(i)
    if i % 3 == 0:
        values.append(li)
        li = []

df = pd.DataFrame(dict(zip(keys, values)))
type(df)
```

```python
# 2차원 배열의 index=, column= 설정가능
# 딕셔너리      -> 데이터프레임 :  시리즈의 합, 즉 데이터가 세로로 들어간다, column명은 키값
# 이차원 리스트 -> 데이터 프레임 : 리스트 하나가 가로데이터... column 명은 따로 지정하지 않으면 0부터 순차적
```

```python
li = []
values = []
for i in range(1,22):
    li.append(i)
    if i % 3 == 0:
        values.append(li)
        li = []

# df = pd.DataFrame(values, index = ['a','b','c','d','e','f','g'], columns = ['A','B','C'])
df = pd.DataFrame(values)
df.index = ['a','b','c','d','e','f','g']
df.columns = ['A','B','C']
# column 이름을 B -> BB로 변경
# df.columns = ['A','BB','C']
df2 = df.rename(columns={
    'B':'BB'
}, inplace = True) # df는 바뀌지 않고 바뀐 변수만 보여주므로, df = 또는 inplace=True 사용
df
# 보통 inplace는 데이터 유지 및 오류 발생 시 다시 가져오기 위해 잘 사용하지 않음
df2 = df.rename(index={
    'f':'A+', 'c':'CC' # dictionary 이용하여 변경
})
df2
```

```python
# drop 기본으로 레코드(행)을 삭제함, axis = 0 (기본으로 행 삭제), axis = 1 (열 삭제)
# 행 삭제... CC, A+
df2.drop(index=['CC', 'A+'])
df3 = df2.drop(['CC', 'A+']) # axis = 0
# df3.drop(columns=['BB'])
df4 = df3.drop(['BB'],axis=1)
df4
```

```python
# 1. 시리즈 만들고 숫자로 연산해 보기
ser = pd.Series([i for i in range(5)])
print(ser)
ser * 5
ser + 1
# 2. 시리즈 두개 만들어서 연산해 보기
ser2 = pd.Series([random.randint(1,10) for i in range(5)],index=[1,2,3,4,5])
print(ser2)
ser + ser2
ser * ser2
ser_df = pd.DataFrame(ser)
ser_df2 = pd.DataFrame(ser2)
ser_df.add(ser_df2, fill_value=0)
```
