```python
temp = df.dtypes
# 연속형 변수 / 범주형 변수
categorical_cols = temp[df.dtypes == 'object'].index
numeric_cols = [col for col in temp.index if col not in categorical_cols]

#
df.select_dtypes(include=['object']).columns
df.select_dtypes(include=['int64', 'float64']).columns
```

```python
# 연속형 수치 데이터
df[numeric_cols]
sequencial_cols = ['age','capital-gain', 'hours-per-week','fnlwgt']
# 연속형 데이터를 카테고리화
df[sequencial_cols[0]].min()
df['age_group'] = pd.cut(
    df[sequencial_cols[0]],
    bins=[0,20,40,80,100],
    labels=['children','young','adult', 'senior']
)
df['capital-gain'].value_counts()
```

```python
import numpy as np
cnt, bins = np.histogram(df['capital-gain'],bins=3) # 3구간으로 나눔
df['capital-gain_group'] = pd.cut(
    df[sequencial_cols[1]],
    bins=bins,
    labels=['low','middle','high']
)
df.head()
```
